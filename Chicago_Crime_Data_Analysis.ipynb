```python
#importing modules
import pandas as pd
import numpy as np
import sqlite3
from pandasql import sqldf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#bringing in chicago crime data
crime_data = pd.read_csv('/Users/chadvonholtz/Desktop/Virginia Tech Fullstack Academy Data Analytics Bootcamp Documents/Tableau/Weekly Projects/Chicago Crime Analysis Project #3/Crimes_-_2001_to_Present.csv')
crime_data
```

```python
#setting max columns to none so that all columns are visible
pd.set_option('display.max_columns', None)
crime_data.head()
```

```python
#checking the dataset
crime_data.info()
```

```python
#subsetting the dataset to look at crime from 2020 and on
crime_data_2 = crime_data.loc[crime_data['Year'] >= 2020]
crime_data_2.info()
```

```python
#checking for duplicates
crime_data_2.duplicated().sum()
```

```python
#duplicates in case numbers
crime_data_2['Case Number'].duplicated().sum()
#110 duplicates of case numbers. Repeat offerenders?
```

```python
#checking null values
crime_data_2.isnull().sum()
```

```python
#changing the casing of the 3 string columns for easier readability
crime_data_2['Block'] = crime_data_2["Block"].str.title()
crime_data_2['Primary Type'] = crime_data_2["Primary Type"].str.title()
crime_data_2['Description'] = crime_data_2["Description"].str.title()
crime_data_2['Location Description'] = crime_data_2["Location Description"].str.title()
crime_data_2.head()
```

```python
#checking to see that the subset includes the correct data in years
crime_data_2.loc[crime_data_2['Year'] == 2020]
```

```python
def domesticarrest(x):
    if (x['Arrest'] == True) & (x['Domestic'] == True):
        return True
    else:
        return False
```

```python
#applying the domestic arrest function to the crime_data_2 df
crime_data_2['Domestic Arrest'] = crime_data_2.apply(domesticarrest, axis=1)
#checking that the function was applied properly
crime_data_2.sample(10)
```

```python
#checking to see that the domestic arrest function worked
crime_data_2.loc[(crime_data_2['Domestic'] == True) & (crime_data_2['Arrest'] == True)]
```

```python
#bringing in csv to map ward and community area
ward_community_map = pd.read_csv('/Users/chadvonholtz/Desktop/Virginia Tech Fullstack Academy Data Analytics Bootcamp Documents/Tableau/Weekly Projects/Chicago Crime Analysis Project #3/CommAreas_202302221912.csv')
ward_community_map
```

```python
#checking data types
ward_community_map.info()
```

```python
#converting area number to float to match crime_data df
ward_community_map['AREA_NUMBE'] = ward_community_map['AREA_NUMBE'].astype(float)
```

```python
#verifying that the column converted to float
ward_community_map.info()
```

```python
#Renaming columns to merge two dataframes
ward_community_map.rename(columns = {'AREA_NUMBE':"Community Area"}, inplace = True)
ward_community_map.rename(columns = {'COMMUNITY' : 'Community'}, inplace = True)
ward_community_map
```

```python
#merging the two dataframes so that customer and store location information are in same df
crime_data_3 = pd.merge(crime_data_2, ward_community_map,how='left', on ='Community Area')
crime_data_3.head()
```

```python
#verifying that the merge worked
crime_data_3.loc[crime_data_3['Community Area'] == 35.0]
```

```python
#Making the Community column more readable
crime_data_3['Community'] = crime_data_3["Community"].str.title()
crime_data_3.head()
```

```python
#sending dataframe to csv to import into Tableau
crime_data_3.to_csv('Chicago_Crime_Data.csv')
```
